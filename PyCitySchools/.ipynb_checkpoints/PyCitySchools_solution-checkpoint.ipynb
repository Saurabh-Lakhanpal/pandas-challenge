# Analysis of PyCity Schools Performance
---
#### Reading Scores
Charter schools in the PyCity Schools district have demonstrated a notable advantage in reading proficiency. The analysis reveals that Charter schools have shown significant gains in reading scores compared to their District school counterparts, even when accounting for similar student demographics.

#### Math Scores
The trend observed in reading scores extends to math performance as well. Charter schools consistently outperform District schools in math, indicating a broader pattern of academic excellence in these institutions.

#### Overall Scores
When considering overall academic performance, Charter schools emerge as clear leaders. The data indicates that a significantly higher percentage of students in Charter schools pass both math and reading tests, with passing rates of 90.43% compared to 53.67% in District schools.

#### Charter Schools
- **Budget**: $7,301,505
- **Student Count**: 12,194
- **Budget per Student**: $598.78

Charter schools, with their smaller student populations, have outperformed public District schools across all metrics. This raises questions about whether the superior performance is due to effective school practices or the benefits of serving smaller student bodies. Further analysis is needed to determine the underlying factors contributing to this success.

#### District Schools
- **Budget**: $17,347,923
- **Student Count**: 26,976
- **Budget per Student**: $643.09

In District schools, smaller and medium-sized institutions have shown significantly better math performance, with passing rates between 89% and 91%, compared to just 67% in larger schools. This suggests that school size may play a critical role in student achievement.

#### Budget Analysis
Despite having a larger total budget, District schools do not significantly outspend Charter schools on a per-student basis. Interestingly, higher overall budgets do not correlate with better test results. In fact, schools with higher spending per student did not necessarily achieve superior academic outcomes, highlighting the need for a more nuanced understanding of how financial resources impact educational performance.

This comprehensive analysis underscores the importance of examining various factors, including school size, budget allocation, and educational practices, to understand the drivers of academic success in PyCity Schools. Further research and targeted interventions may help replicate the success of Charter schools across the district.

# <ins>CODE & ANALYSIS STEPS</ins>

## Data Sample

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path


# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")
school_data_complete.head(10)
```

# <ins>DISTRICT SUMMARY</ins>

```python
# Calculate the total number of unique schools

school_count = len(school_data_complete["school_name"].unique())
print(f"The total number of given schools in the school board data are", school_count)
```

```python
# Calculate the total number of students

student_count = len(school_data_complete["Student ID"].unique())
print(f"The total number of students given in the school board data are", f"{student_count:,}")
```

```python
# Calculate the total budget

total_budget = school_data["budget"].sum()
print(f"The total of budget allocated for all the 15 schools of the school board is", f"${total_budget:,}")
```

```python
# Calculate the average (mean) math score

average_math_score = student_data["math_score"].mean()
print(f"The Average of Maths scores among all of the", f"{student_count:,}", "given student through the school board is", round(average_math_score, 2))
```

```python
# Calculate the average (mean) reading score

average_reading_score = student_data["reading_score"].mean()
print(f"The Average of Maths scores among all of the", f"{student_count:,}", "given student through the school board is", round(average_reading_score, 2))
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)

passing_score = 70

passing_math_count = school_data_complete[(school_data_complete["math_score"] >= passing_score)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100

print(f"Among the", f"{student_count:,}", "given student through the school board,", f"{round(passing_math_percentage, 2)} %", "scored above 70 marks in Maths")
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)

passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= passing_score)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100

print(f"Among the", f"{student_count:,}", "given student through the school board,", f"{round(passing_reading_percentage, 2)} %", "scored above 70 marks in reading")
```

```python
# Use the following to calculate the percentage of students that passed math and reading

passing_math_reading_count = school_data_complete[(school_data_complete["math_score"] >= passing_score) \
                                                  & (school_data_complete["reading_score"] >= passing_score)].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100

print(f"Among the", f"{student_count:,}", "given student through the school board,", f"{round(overall_passing_rate, 2)} %", "scored above 70 marks in both maths & reading")
```

## Snapshot of District's Key Metrics

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary_series = [{"Total Schools": school_count, 
                            "Total Students": student_count, 
                            "Total Budget": total_budget, 
                            "Average Math Score":  round(average_math_score,2), 
                            "Average Reading Score":  round(average_reading_score,2), 
                            "% Passing Math": round(passing_math_percentage,2),
                            "% Passing Reading": round(passing_reading_percentage,2),
                            "% Overall Passing Rate": round(overall_passing_rate,2)
                            }]

district_summary = pd.DataFrame(district_summary_series)

# Formatting

district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.2f}%".format)
district_summary["% Overall Passing Rate"] = district_summary["% Overall Passing Rate"].map("{:,.2f}%".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)

# Display the DataFrame
district_summary
```

# <ins>SCHOOL SUMMARY</ins>

```python
# Use the code provided to select the type per school from school_data
school_types = school_data.set_index(["school_name"])["type"].sort_index().sort_values()

```

```python
# Calculate the total student count per school from school_data
per_school_counts = student_data.value_counts("school_name").sort_index()
```

```python
# Calculate the total school budget from school_data

per_school_budget = school_data.set_index(["school_name"])["budget"].sort_index()
```

```python
# Calculate the per capita spending per student from school_data

per_school_capita = per_school_budget/per_school_counts
```

```python
# Calculate the average maths scores per school from school_data_complete


per_school_math = school_data_complete.groupby("school_name")["math_score"].mean()
per_school_reading = school_data_complete.groupby("school_name")["reading_score"].mean()
```

```python
# Calculate the number of students per school with math scores of 70 or higher from school_data_complete

school_students_passing_math = school_data_complete[school_data_complete["math_score"] >= passing_score].groupby("school_name").size().sort_values()
```

```python
# Calculate the number of students per school with reading scores of 70 or higher from school_data_complete

school_students_passing_reading= school_data_complete[school_data_complete["reading_score"] >= passing_score].groupby("school_name").size().sort_values()
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= passing_score)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size().sort_values()
```

## Passing Score Summary

```python
# Create a DataFrame called `passing_score_summary` with columns for the calculations above.
passing_score_summary = pd.DataFrame({  "Average Passing Maths Score"           : school_students_passing_math,
                                        "Average Passing Reading Score"         : school_students_passing_reading,
                                        "Average Passing Maths & Reading Score" : school_students_passing_math_and_reading,
                                    })

passing_score_summary
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_students_passing_math / per_school_counts * 100
per_school_passing_reading = school_students_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100


passing_rates_dict = {  "Per School Math Passing %": per_school_passing_math, 
                        "Per School Reading Passing %": per_school_passing_reading, 
                        "Per School Overall Passing %": overall_passing_rate
                        }

passing_rates = pd.DataFrame(passing_rates_dict).sort_values(["Per School Overall Passing %"])

```

## School Data Summary

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({ "School Type"           : school_types,
                                    "Total Students"        : per_school_counts,
                                    "Total School Budget"   : per_school_budget,
                                    "Per Student Budget"    : per_school_capita,
                                    "Average Math Score"    : per_school_math,
                                    "Average Reading Score" : per_school_reading,
                                    "% Passing Math"        : per_school_passing_math,
                                    "% Passing Reading"     : per_school_passing_reading,
                                    "% Passing Overall"     : overall_passing_rate
                                    })

per_school_summary_formatted = per_school_summary.copy()

# Formatting
per_school_summary_formatted["Total School Budget"] = per_school_summary_formatted["Total School Budget"].map("${:,.2f}".format)
per_school_summary_formatted["Per Student Budget"] = per_school_summary_formatted["Per Student Budget"].map("${:,.2f}".format) 
per_school_summary_formatted["Average Math Score"] = per_school_summary_formatted["Average Math Score"].map("{:,.2f}".format)
per_school_summary_formatted["Average Reading Score"] = per_school_summary_formatted["Average Reading Score"].map("{:,.2f}".format)
per_school_summary_formatted["% Passing Math"] = per_school_summary_formatted["% Passing Math"].map("{:,.2f}%".format)
per_school_summary_formatted["% Passing Reading"] = per_school_summary_formatted["% Passing Reading"].map("{:,.2f}%".format)
per_school_summary_formatted["% Passing Overall"] = per_school_summary_formatted["% Passing Overall"].map("{:,.2f}%".format)

# Display the DataFrame
per_school_summary_formatted
```

## Highest-Performing Schools (by % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.
top_schools = round(per_school_summary.sort_values(["% Passing Overall"], ascending=False),2)

top_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.
bottom_schools = round(per_school_summary.sort_values(["% Passing Overall"], ascending=True))
bottom_schools.head(5)
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders.groupby("school_name").mean(numeric_only=True)["math_score"]
tenth_grader_math_scores = tenth_graders.groupby("school_name").mean(numeric_only=True)["math_score"]
eleventh_grader_math_scores = eleventh_graders.groupby("school_name").mean(numeric_only=True)["math_score"]
twelfth_grader_math_scores = twelfth_graders.groupby("school_name").mean(numeric_only=True)["math_score"]

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({
                                    "9th"   : round(ninth_grade_math_scores,2),
                                    "10th"  : round(tenth_grader_math_scores,2),
                                    "11th"  : round(eleventh_grader_math_scores,2),
                                    "12th"  : round(twelfth_grader_math_scores,2)
                                    })

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

## Reading Score by Grade 

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.

ninth_grade_reading_scores = ninth_graders.groupby("school_name").mean(numeric_only=True)["reading_score"]
tenth_grader_reading_scores = tenth_graders.groupby("school_name").mean(numeric_only=True)["reading_score"]
eleventh_grader_reading_scores = eleventh_graders.groupby("school_name").mean(numeric_only=True)["reading_score"]
twelfth_grader_reading_scores = twelfth_graders.groupby("school_name").mean(numeric_only=True)["reading_score"]

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.DataFrame({
                                    "9th"   : round(ninth_grade_reading_scores,2),
                                    "10th"  : round(tenth_grader_reading_scores,2),
                                    "11th"  : round(eleventh_grader_reading_scores,2),
                                    "12th"  : round(twelfth_grader_reading_scores,2)
                                    })


# Minor data wrangling
reading_scores_by_grade = reading_scores_by_grade[["9th", "10th", "11th", "12th"]]
reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary for later aggregations
school_spending_df = round(per_school_summary.copy(),2)
```

```python
# Use `pd.cut` on the per_school_capita Series from earlier to categorize per student spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_summary["Per Student Budget"], spending_bins, labels=spending_labels) 

# Convert Spending Ranges (Per Student) to a string
school_spending_df["Spending Ranges (Per Student)"] = school_spending_df["Spending Ranges (Per Student)"].astype(str)
school_spending_df
```

```python
#  Calculate averages for the desired columns.

spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean(numeric_only=True)
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean(numeric_only=True)
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean(numeric_only=True)
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean(numeric_only=True)
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Overall"].mean(numeric_only=True)
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({"Average Math Score": round(spending_math_scores, 2),
                                "Average Reading Score": round(spending_reading_scores,2),
                                "% Passing Math": round(spending_passing_math,2),
                                "% Passing Reading": round(spending_passing_reading,2),
                                "% Overall Passing": round(overall_passing_spending,2)
                                })

# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Create a copy of the school summary for later aggregations
school_size_df = per_school_summary.copy()
```

```python
# Use `pd.cut` on the per_school_counts Series from earlier to categorize school size based on the bins.
school_size_df["School Size"] = pd.cut(per_school_summary["Total Students"], size_bins, labels=size_labels)

# Convert School Size to a string
school_size_df["School Size"] = school_size_df["School Size"].astype(str)
round(school_size_df,2)
```

```python
# Calculate averages for the desired columns.
size_math_scores = school_size_df.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = school_size_df.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_size_df.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = school_size_df.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = school_size_df.groupby(["School Size"])["% Passing Overall"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({"Average Math Score": round(size_math_scores,2),
                             "Average Reading Score": round(size_reading_scores,2),
                             "% Passing Math": round(size_passing_math,2),
                             "% Passing Reading":round(size_passing_reading,2),
                             "% Passing Overall": round(size_overall_passing,2)
                             })

# Display results
size_summary
```
